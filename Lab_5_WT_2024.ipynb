# Factor Escala $a$: slo interpretar.

Dada la seal: $x(t)=\sin(6 \pi t)$, con $t \in [0,4)$.

Que pasa al realizar un cambio de escala de factor $a=2$ y $a=\frac{1}{2}$

es decir se generan las seales $x_2(t)=x(\frac{t}{2})$ y $x_{0.5}(t)=x(\frac{t}{1/2})$

Muestreamos $x$ y $x_2$ con perodo de muestreo $\frac{1}{64}$.

```python
pi= np.pi
PM=1/64#perodo de muestreo
FM=64# 1/PM, frecuencia de muestreo en Hertz
t=np.arange(0,4,1/64)
x=np.sin(6*pi*t)
x2=np.sin(3*pi*t)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(t,x, color='blue')
plt.title('Seal Original')

plt.subplot(212)
plt.plot(t,x2, color='green')
plt.title('Seal escala a=2')
plt.xlabel('tiempo muestreado')
plt.show()
```

## Sub-muestreo y sobre-muestreo

```python
cant_m=len(x)
print(cant_m)

x_sob=signal.resample(x,cant_m*2)
x_sub=signal.resample(x,cant_m//2)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
#plt.subplot( sharex=True)
plt.plot(x_sob, color='green', label='Senal sobre muestreada')
plt.plot(x, color='blue', label='Senal Original')
plt.legend(loc='upper right')

#plt.subplot(212)
#plt.plot(x, color='blue', label='Senal Original')
plt.plot(x_sub, color='red', label='Senal sub muestreada')
plt.legend(loc='upper right')
plt.show()
print('comparar las seales y la cantidad de muestras')
```

mayor escala, se dilata, baja frecuencia (anlogo a sobremuestear)

# Transformada Ondita (Contnua y Discreta)

- Transformada Ondita Continua  (cwt o CWT por sus siglas en ingls),
- Transformada Ondita Discreta (dwt o DWT) no tiene redundancias y es ms adecuada para el estudio digital.


 En el abordaje con la transformada Ondita se busca descomponer la seal como suma de seales de una base Ortonormal formada por una ondita madre $\psi$ junto con escalas y traslaciones de sta: $\psi_{a,b}$ ($a$ escala, $b$ traslacin):

 $x(t)= \sum_{a,b} C_{a,b}\, \psi_{a,b}(t)$

 Donde los coeficientes son $C_{a,b}=\left \langle x(t),\psi_{a,b}(t) \right \rangle$.

El PI(producto interno) nos dice que tan parecida es la seal $x(t)$ a la seal de la base y su aporte en la descomposicin.

# CWT: Transformada Ondita Continua

Transformada Ondita Continua  (cwt o CWT por sus siglas en ingls)


Analicemos la seal Chirp lineal usando funcin pywt.cwt

```python
t = np.arange(0, 2.00, 0.001) #tiempo con 1000 Hz frec  de muestreo (Perodo o paso 0.001s)
C1=signal.chirp(t,30,1,60) #frec inicial 30Hz y 60Hz al segundo
delta_t = 0.001  #perodo de muestreo, sampling_period

max_escala = 300 #p/ escalas
scales = np.arange(6,max_escala,2)
wavelet_type = 'morl'

coefs, freqs = pywt.cwt(C1, scales, wavelet_type, delta_t)

print(coefs.shape)
print(scales.shape)
print(freqs.shape)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(311)
plt.plot(t,C1)
plt.xlabel('tiempo en seg. muestreado a 1000 Hrtz')
plt.title('Chirp Lineal')

plt.subplot(312)
plt.pcolormesh(t,freqs,coefs)
plt.xlabel('tiempo')
plt.ylabel('frecuencia') #??
plt.colorbar()

plt.show()
print('notar multiresolucin;' )
print('a baja frecuencia (30Hz aprox, en los inicios de la seal) ver Deltaw y deltat')
print('al avanzar en el tiempo sube la frecuencia (70-90 Hrts) ver deltaw y deltat')
```

### Filtros: interpretacin

Diseamos un filtro ortogonal de Daubechies pasa-bajo de mnima fase de longitud 8 y encontramos el filtro pasa-alto correspondiente.
Graficamos la magnitud (y fase) de ambos filtros. Observe y complete verifique


```python
#primero encontramos los coeficientes del filtro
#si la longitud deseada es 8=2*p, entonces el nivel de la ondita debe ser p=8/2. O sea dB4.

h=daub(4) #asociado a la escala, pasa-bajo
g=qmf(h) #asociado a la ondita, pasa-alto

#haremos convolucin con estos coeficientes (filtraremos con ellos)
#Veamos como se comportan en frecuencia

#Transformamos y centramos
N = 1024 #longitud de un posible segmento o seal para analizar (seal o segmento con N muestras)
H = fft(h, N) #DFT punto N, para estudiar el efecto del filtro en frecuencia,

G  = fft(g, N)
HC = fftshift(H) #centramos
GC = fftshift(G) #centramos
```

```python
#Graficamos mdulo de las transformadas

frecc = np.arange(-np.pi, np.pi, 2*np.pi/N) #frec norm. y centradas en 0, van en [-pi, pi)

plt.figure(figsize=(8,3))
plt.subplot(121)
plt.plot(frecc, np.abs(HC/max(HC)), color='blue', label='Mdulo de H: Pasa ...') #completar
plt.legend(loc='upper right')

plt.subplot(122)
plt.plot(frecc, np.abs(GC/max(GC)), color='red', label='Mdulo de G: Pasa ...') #completar
plt.legend(loc='upper right')
plt.xlabel('Frecuencia en [$-\pi$,$\pi$)')
plt.show()
```

**Descomponemos la seal de audio 'audiolab1.wav en aproximaciones y detalles utilizando la Transformada Ondita Discreta con ondita "haar"**

```python
au2=read('audiolab1.wav')
print (au2) #terna
matriz=au2[1] #matriz de datos, con dos colmnas, una por cada canal
A2=matriz[:,0] #primer canal columna 0, vector de datos
print('matriz con dos columnas:', matriz)
print('vector:',A2)

plt.figure(figsize=(20,4))
plt.plot(A2, color='red')
plt.title("Seal de audio, canal 0")
plt.show()
```

```python
#calculo la transformada ondita discreta con dwt con ondita haar
#a[n]=(x*h)[2n] conv filtro pasa bajo y submuestreo
#d[n]=(x*g)[2n] conv filtro pasa alto  y submuestreo

#Aqu cA es la aproximacin (baja frecuencia) y d es el detalle (alta frecuencia)
cA_h, cD_h = pywt.dwt(A2, 'haar')

plt.figure(figsize=(10,8))
plt.subplot(311)
plt.plot(A2, color='blue', label='Audio, frec 0-22050 Hz (Nyquist)')
plt.legend(loc='upper right')
plt.subplot(312)
plt.plot(cA_h, color='red', label='Aproximacin, frec. bajas: 0-11024 Hz')
plt.legend(loc='upper right')
plt.subplot(313)
plt.plot(cD_h, color='green', label='Detalles, frec. altas: 11025-22050 Hz')
plt.legend(loc='upper right')
plt.show()
print('NOTAR LAS DIFERENCIAS DE LONGITUD')
print('Notar que los Detalles y Alta Frecuencia estn asociados al ruido, suelen tener menor amplitud')
```

**Descomponemos la seal en aproximaciones y detalles utilizando la Transformada Ondita Discreta con ondita "symlets"**

```python
cA_s, cD_s = pywt.dwt(A2, 'sym2')

plt.figure(figsize=(10,8))
plt.subplot(311)
plt.plot(A2, color='blue', label='Audio, frec 0-22050 Hz (Nyquist)')
plt.legend(loc='upper right')
plt.subplot(312)
plt.plot(cA_s, color='red', label='Aproximacin, frec. bajas: 0-11024 Hz')
plt.legend(loc='upper right')
plt.subplot(313)
plt.plot(cD_s, color='green', label='Detalles, frec. altas: 11025-22050 Hz')
plt.legend(loc='upper right')
plt.show()
```
