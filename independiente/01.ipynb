## Ej. 6
Sea la señal:
$x(t)=5.sen(6 \pi t)+cos(14\pi)$

a) Que componentes de frecuencias (de oscilación) tiene la señal en Hertz? que amplitud tiene cada uno?

La señal tiene los siguiente compoenentes en frecuencia:
- $f_1 = 3[Hz]$
- $f_2 = 7 [Hz]$

b) Digitalice y grafique la señal $x$ con 1000 muestras con $t$ (en segundos) en el intervalo $[0,4)$. Calcule la frecuencia de muestreo, período de muestreo y Frecuencia de Nyquist del muestreo (Fm, Pm, Nm)

```python
M = 1000 # cantidad de muestras solcitadas
t = np.linspace(0,4,M, endpoint=False) # vector de tiempos discretizados
# señal digitalizada
x= 5*np.sin(6*np.pi*t)+np.cos(14*np.pi*t)

# frecuencia de muestreo
l = 4 # longitud del intervalo
Fm = M/l
print(f"Frecuencia de muestreo: {Fm}")

Pm = 1/Fm
print(f"Período de muestreo: {Pm}")

Nm = Fm/2
print(f"Nyquist de muestreo: {Nm}")
```

## Ej. 8

Realice un primer análisis de frecuencias de la señal:

$g(t)=7.sen(10\pi t)-8.cos(14\pi t)+ (1/2).sen(50\pi t)$

Digitalícela con 1000 muestras y período de muestreo Pm= 0.01. Cuanto tiempo dura la señal?
Grafique la señal muestreada en función del tiempo y luego en función de las muestras.

```python
M = 1000 # cantidad de muestras
Pm = 0.01 # período de muestreo
Fm = 1/Pm # frecuencia de muestreo
t = M/Fm # longitud del intervalo
print(f"Duración de la señal: {t} segundos")
# defino el vector de tiempos 
vt = np.linspace(0,t, M)
# defino la señal
g1 = 7*np.sin(10*np.pi*vt)
g2 = -8*np.cos(14*np.pi*vt)
g3 = 0.5*np.sin(50*np.pi*vt)
g = g1+g2+g3

# grafico la señal en función del vector de tiempos discretizados
plt.figure(figsize=(10,5))
plt.plot(vt,g, color="green")
plt.title("Señal g(t) vs vector tiempos")
plt.xlabel("Vector de tiempos discretizados [s]")
plt.ylabel("Valor de g(t)")
plt.grid()
# grafico la señal en función de las muestras
plt.figure(figsize=(10,5))
plt.stem(g)
plt.title("Señal g(t) vs muestras")
plt.xlabel("Muestras de g(t)")
plt.ylabel("Valor de g(t) en la muestra")

plt.grid()
plt.show()
```
